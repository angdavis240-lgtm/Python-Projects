```python
#Import python libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import the data
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/refs/heads/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
#print(df.head())
#print(df.info())
#print(df.describe())
df.columns
df.columns = df.columns.str.strip()

```

```python
#School for initial comparison point is East Side Community School vs all Manhattan schools
#Selected feature are 'Mean Score' and 'Percent Scoring 80 or Above'

#df['School Name'].isnull().sum()
#df['Mean Score'].isnull().sum()
#df['Percent Scoring 80 or Above'].isnull().sum()

#There are no null values
```

```python
#Sample DataFrame with 's' values for Mean Score values
df = pd.DataFrame({'Mean Score':['s', 77.9, 74, 's', 's', 67.4]})
print(df)
#Replace 's' with Nan values
df['Mean Score'] = df['Mean Score'].replace('s', np.nan)
#Convert the column to a numeric type
df['Mean Score'] = pd.to_numeric(df['Mean Score'], errors='coerce')
# Calculate the median of the numeric values
median_value = df['Mean Score'].median()
print(f"\nCalculated Median: {median_value}")
# Fill NaN values with the median
df['Mean Score'] = df['Mean Score'].fillna(median_value)

print("\nDataFrame after converting 's' to median:")
print(df)
```

```python
#Sample DataFrame with 's' values for 'Percent Scoring 80 or Above' values
df = pd.DataFrame({'Percent Scoring 80 or Above':['s',43.8,22.2,'s','s',0,]})
print(df)
# Replace 's' with NaN values
df['Percent Scoring 80 or Above']= df['Percent Scoring 80 or Above'].replace('s', np.nan)
# Convert to numeric type
df['Percent Scoring 80 or Above']= pd.to_numeric(df['Percent Scoring 80 or Above'], errors= 'coerce')
print(df)
# Calculate the mean of the numeric values
mean_value = df['Percent Scoring 80 or Above'].mean()
print("f\nCalculated Mean: {mean_value}")
# Fill NaN values with the mean
df['Percent Scoring 80 or Above'] = df['Percent Scoring 80 or Above'].fillna(mean_value)
print("\nDataFrame after converting 's' to mean")
print(df)
```

```python
df.columns = df.columns.str.strip()
df['School Name'] = df['School Name'].astype(str).str.strip()
east_side = df[df['School Name'].str.contains('East Side Community School', na=False)]
df['School DBN'] = df['School DBN'].astype(str)
manhattan = df[df['School DBN'].str.startswith('M', na=False)]
print("East Side Community School:")
print(east_side)

print("\nManhattan Schools:")
print(manhattan)
```

```python
# Compare descriptive statistics of East Side Community School vs all Manhattan schools
east_side['Percent Scoring 80 or Above'] = pd.to_numeric(
    east_side['Percent Scoring 80 or Above'],errors='coerce'
)
manhattan['Percent Scoring 80 or Above'] = pd.to_numeric(
    manhattan['Percent Scoring 80 or Above'], errors='coerce'
)
east_side['Mean Score'] = pd.to_numeric(
    east_side['Mean Score'], errors='coerce'
)
manhattan['Mean Score'] = pd.to_numeric(
    manhattan['Mean Score'], errors='coerce'
)
print(manhattan['Percent Scoring 80 or Above'].isna().sum()) # Number of NaNs = 0
print(manhattan['Percent Scoring 80 or Above'].count()) # Number of valid numeric values = 0

# Replace 's' or blanks with NaN
manhattan['Percent Scoring 80 or Above'] = manhattan['Percent Scoring 80 or Above'].replace('s', np.nan)
manhattan['Mean Score'] = manhattan['Mean Score'].replace('s', np.nan)
#Convert to numeric again
manhattan['Percent Scoring 80 or Above'] = manhattan['Percent Scoring 80 or Above'].astype(str).str.strip()
manhattan['Percent Scoring 80 or Above'] = manhattan['Percent Scoring 80 or Above'].str.replace(r'[\u00A0]', '', regex=True)
placeholders = ['s', '-', 'N/A', '']
manhattan['Percent Scoring 80 or Above'] = pd.to_numeric(manhattan['Percent Scoring 80 or Above'], errors='coerce')
manhattan['Mean Score'] = manhattan['Mean Score'].astype(str).str.strip()
manhattan['Mean Score'] = manhattan['Mean Score'].str.replace(r'[\u00A0]', '', regex= True)
placeholders = ['s', '-', 'N/A', '']
manhattan['Mean Score'] = pd.to_numeric(manhattan['Mean Score'], errors= 'coerce')
# East Side Community School
east_mean = east_side['Percent Scoring 80 or Above'].mean()
east_median = east_side['Percent Scoring 80 or Above'].median()
east_mode = east_side['Percent Scoring 80 or Above'].mode()
east_mean_ms = east_side['Mean Score'].mean()
east_median_ms = east_side['Mean Score'].median()
east_mode_ms = east_side['Mean Score'].mode()

# All Manhattan Schools
columns = ['Percent Scoring 80 or Above', 'Mean Score']

for c in columns:
    manhattan[c] = manhattan[c].astype(str).str.replace(r'[\u00A0,%]', '', regex=True).str.strip()
    manhattan[c] = manhattan[c].replace(['s','-','N/A',''], np.nan)
    manhattan[c] = pd.to_numeric(manhattan[c], errors='coerce')
print("After cleaning/conversion â€” sample values:")
print(manhattan[columns].head(10))
manhattan_mean = manhattan['Percent Scoring 80 or Above'].mean()
manhattan_median = manhattan['Percent Scoring 80 or Above'].median()
manhattan_mode = manhattan['Percent Scoring 80 or Above'].mode()
manhattan_mean_ms = manhattan['Mean Score'].mean()
manhattan_median_ms = manhattan['Mean Score'].median()
manhattan_mode_ms = manhattan['Mean Score'].mode()
print(f"East Side Community School(Percent Scoring 80 or Above) - Mean: {east_mean}, Median: {east_median}, Mode: {east_mode}")
print(f"All Manhattan Schools(Percent Scoring 80 or Above) - Mean: {manhattan_mean}, Median: {manhattan_median}, Mode: {manhattan_mode}")
print(f"East Side Community School(Mean Score) - Mean: {east_mean_ms}, Median: {east_median_ms}, Mode: {east_mode_ms}")
print(f"All Manhattan Schools (Mean Score) - Mean: {manhattan_mean_ms}, Median: {manhattan_median_ms}, Mode: {manhattan_mode_ms}")

```

```python
metrics = {
    "Schools": ["East Side Community School", "East Side Community School", "Manhattan Schools", "Manhattan Schools"],
    "Metric": ["Mean Score", "Percent >= 80", "Mean Score", "Percent >= 80"],
    "Value": [
        east_side["Mean Score"].mean(),
        east_side["Percent Scoring 80 or Above"].mean(),
        manhattan["Mean Score"].mean(),
        manhattan["Percent Scoring 80 or Above"].mean()
    ]
}

df_compare = pd.DataFrame(metrics)

plt.figure(figsize=(10, 6))
sns.barplot(data=df_compare, x="Metric", y="Value", hue="Schools")
plt.title("East Side Community School vs All Manhattan Schools")
plt.ylabel("Score")
plt.xlabel(" ")
plt.legend(title="Schools")
plt.tight_layout()
plt.show()
```
