```python
# Importing python libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Plot settings
plt.style.use('default')
sns.set(style='whitegrid')

# Showing versions
import sys
print('Python version:', sys.version)
print('Pandas version:', pd.__version__)
print('Seaborn version:', sns.__version__)
```

```python
# Loading of NYPD Shooting Incident Data (Year to Date). Dataset focuses on the current calendar year (2025)
from google.colab import files

uploaded = files.upload()
file_name = list(uploaded.keys())[0]
print('Uploaded file:', file_name)

df = pd.read_csv(file_name)
```

```python
# Peak at the data
print('First 5 rows:')
display(df.head())

print('Last 5 rows:')
display(df.tail())

# Shape of dataset
print('Shape (rows, columns):', df.shape)

# Column names
print('\nColumn names:')
print(df.columns.tolist())

# Data types
print('\nData types:')
print(df.dtypes)

# Info (non-null counts, dtypes)
print('\nDataFrame info:')
display(df.info())

# Basic numeric stats
print('\nDescriptive statistics (numeric columns):')
display(df.describe())

# Basic stats for categorical columns
print('\nDescriptive statistics (categorical columns):')
display(df.describe(include='object'))

# Memory usage
print('\nMemory usage (bytes):')
print(df.memory_usage(deep=True))
```

```python
# Count of missing values
print('Missing values per column (count):')
missing_count = df.isnull().sum()
display(missing_count)

# Percent of missing values
print('\nMissing values per column (percentage):')
missing_pct = df.isnull().mean()*100
display(missing_pct)

# Simple table sorted by percentage
print('\nColumns sorted by missing percentage:')
display(missing_pct.sort_values(ascending=False))

# List number of unique values per column
print('\nNumber of unique values per column:')
display(df.nunique())

# Visualization: missing values heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False)
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Number of duplicate rows
num_duplicates = df.duplicated().sum()
print('Number of duplicate rows:', num_duplicates)

# Show duplicate rows
if num_duplicates > 0:
  print('\nDuplicate rows:')
  display(df[df.duplicated()])
else:
  print('No duplicate rows found.')

# Create a version without duplicates (SAVE FOR LATER (JUST IN CASE))
df_no_dupes = df.drop_duplicates().copy()
print('Shape after removing duplicates:', df_no_dupes.shape)
```

```python
# Removed spaces from the beginning and ending of column names, lower case, replaced spaces with underscores, replaced - with underscores
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('-','_')
    )
df.columns
```

```python
# used a for loop to convert the displayed columns that were listed as object data type to the category data type.
categorical_cols = ['occur_date', 'occur_time', 'boro', 'loc_of_occur_desc',
       'loc_classfctn_desc', 'location_desc',
       'statistical_murder_flag', 'perp_age_group', 'perp_sex', 'perp_race',
       'vic_age_group', 'vic_sex', 'vic_race', 'x_coord_cd', 'y_coord_cd','new_georeferenced_column'
]

for col in categorical_cols:
  df[col] = df[col].astype('category')

df[categorical_cols].dtypes
```

```python
# I'm going to be comparing the Bronx and Staten Island, I created their respective DataFrames for comparison purposes.
bronx_df = df[df['boro'] == 'BRONX']
staten_island_df = df[df['boro'] == 'STATEN ISLAND']
# BRONX VS. STATEN ISLAND PERP AGE GROUP
# displays summary of the basic statistics for the perp age group in the Bronx
print('\nPerp_Age_Group in the Bronx Summary:')
print(bronx_df['perp_age_group'].describe()) # This shows that the perp age group is typically unknown in the Bronx and this age group appeared 93 times in the dataset which shows that they committed 93 shooting incidents in that boro.


# displays summary of the basic statistics for the perp age in Staten Island
print('\nPerp_Age_Group in Staten Island Summary:')
print(staten_island_df['perp_age_group'].describe()) # This shows that the perp age is typically unknown in Staten Island and this age group committed 5 shooting incidents in that boro.

```

```python
# displays summary of the basic statistics for the perp race in the Bronx
print('\nPerp_Race in the Bronx Summary:')
print(bronx_df['perp_race'].describe()) # This shows that the most frequently recorded perp race category is Black, displaying in 123 shooting incidents in the Bronx.



# displays summary of the basic statistics for the perp race in Staten Island
print('\nPerp_Race in Staten Island Summary:')
print(staten_island_df['perp_race'].describe()) # This shows that the most frequently recorded perp race category is Unknown, displaying in 5 shooting incidents in Staten Island.
```

```python
# Visualization
bronx_race_counts = bronx_df['perp_race'].value_counts()
staten_race_counts = staten_island_df['perp_race'].value_counts()

race_comparison = pd.DataFrame({
    'Bronx': bronx_race_counts,
    'Staten Island': staten_race_counts
}).fillna(0)

race_comparison.plot(kind='bar', figsize=(10,6))
plt.title('Perpetrator Race Comparison: Bronx vs Staten Island')
plt.xlabel('Perp_Race')
plt.ylabel('Number of Shooting Incidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

In conclusion, the Bronx had a significantly higher number of shooting incidents, with a perpretrator's race listed  most frequently as Black, whereas, Staten Island had fewer shooting incidents with the perpetrator's race being frequently listed as unknown. In addition, both, Staten Island and the Bronx had significantly high rates of unknown ages of the perpetrators. The bias displayed in this dataset was the large amounts of unknown values which affected the demographic patterns that were attempting to be displayed in the dataset.
