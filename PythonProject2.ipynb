```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.style.use('default')
sns.set(style='whitegrid')

import sys
print('Python version:', sys.version)
print('Pandas version:', pd.__version__)
print('Seaborn version:', sns.__version__)
```

```python
# Loading of diabetes dataset.
df = pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv')
```

```python
# Peak at the data
print('First 5 rows:')
display(df.head())
print('Last 5 rows:')
display(df.tail())

# Shape of dataset
print('\nShape (rows, columns):', df.shape)

# Column names
print('\nColumn names:')
display(df.columns.tolist())

# Data types
print('\nData Types:')
display(df.dtypes)

# Info (non-null counts, dtypes)
print('\nDataFrame Info:')
display(df.info())

# Basic numeric stats
print('\nDescriptive statistics (numeric columns):')
display(df.describe())

# Basic stats for categorical columns
print('\nDescriptive statistics (categorical columns):')
display(df.describe(include='object'))

# Memory usage
print('\nMemory Usage (bytes):')
display(df.memory_usage(deep=True))
```

```python
# Count of missing values
print('Missing values per column (count):')
missing_count = df.isnull().sum()
display(missing_count)

# Percentage of missing values
print('\nMissing values per column (percentage):')
missing_pct = df.isnull().mean()*100
display(missing_pct)

#Simple table sorted by percentage
print('\nColumns sorted by missing percentage:')
display(missing_pct.sort_values(ascending=False))

# Count unique values per column
print('\nNumber of unique values per column:')
print(df.nunique())

#Visual: missing heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(),cbar=False)
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Number of duplicate rows
num_duplicates = df.duplicated().sum()
print('Number of duplicate rows:', num_duplicates)

#Show duplicate rows (if any)
if num_duplicates > 0:
  print('\nDuplicate rows:')
  display(df[df.duplicated()])
else:
  print('No duplicate rows found.')
```

```python
# Removed any possible spaces in the front and back of the column names, ensured that all column names are lower case, replaced spaces with undescores and replaced - with underscores to improve readability.
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('-', '_')
)
df.columns
```

```python
# Displays random sample of the dataset
df.sample(5, random_state= 29)
```

```python
# Viewed unique values in time in hospital column
df['time_in_hospital'].unique()[:20]
```

```python
# Counts the amount of occurences of each unique value within the time in hospital column.
counts = df['time_in_hospital'].value_counts()
print(counts)
```

```python
#Viewing unique values in age column
df['age'].unique()[:20]
```

```python
# Extracted the midpoint of the age range for the age to be recognized as numeric which created a new column.
age_mapping = {
    '[0-10)':5,
    '[10-20)':15,
    '[20-30)':25,
    '[30-40)':35,
    '[40-50)':45,
    '[50-60)':55,
    '[60-70)':65,
    '[70-80)':75,
    '[80-90)':85,
    '[90-100)':95
}
df['age_numeric']= df['age'].astype(str).map(age_mapping)
print(df['age_numeric'].head())

```

```python
#Viewing unique values in gender column
df['gender'].unique()[:20]
```

```python
#There was 'Unknown/Invalid' value in the 'gender' column that was displayed in the unique values, I removed it by asking to keep only male and female values but it stayed there and was displaying NaN values then when I converted the datatype to string.

print(f"Before:{len(df)} rows")
df = df[df['gender'].isin(['Female', 'Male'])]
df['gender'] = df['gender'].astype(str)
print(f"After: {len(df)} rows")
```

```python
# used a for loop to convert the displayed columns that were listed as object data type to the category data type.
categorical_cols = [
    'race', 'gender', 'age', 'weight', 'payer_code', 'medical_specialty', 'diag_1', 'diag_2', 'diag_3', 'max_glu_serum',
    'a1cresult', 'metformin', 'repaglinide', 'nateglinide', 'chlorpropamide', 'glimepiride', 'acetohexamide',
    'glipizide', 'glyburide', 'tolbutamide', 'pioglitazone', 'rosiglitazone', 'acarbose', 'miglitol',
    'troglitazone', 'tolazamide', 'examide', 'citoglipton', 'insulin', 'glyburide_metformin', 'glipizide_metformin',
    'glimepiride_pioglitazone', 'metformin_rosiglitazone', 'metformin_pioglitazone', 'change', 'diabetesmed', 'readmitted'
]

for col in categorical_cols:
  df[col] = df[col].astype('category')

df[categorical_cols].dtypes
```

```python
# Numeric columns
num_cols = df.select_dtypes(include=['int64']).columns.tolist()
print('Numeric columns:', num_cols)
```

```python
# Displays summary of the basic statistics for age numeric column and time in hospital column.
print('\nAge Numeric Summary:')
print(df['age_numeric'].describe())

print('\nTime In Hospital Summary:')
print(df['time_in_hospital'].describe())
```

```python
# Descriptive statistics
#Displays summary of the basic statistics of the amount of time (in days) that each age number spent in the hospital
print(df.groupby('age_numeric')['time_in_hospital'].describe())

print(df.groupby('age')['time_in_hospital'].describe())
# Displays summary of the basic statistics of the amount of time (in days) that each gender spent in the hospital.
print(df.groupby('gender')['time_in_hospital'].describe())
```

```python
def show_category_distribution(col_name):
  counts = df[col_name].value_counts()
  percents = df[col_name].value_counts(normalize=True)*100
  display(pd.DataFrame({'count': counts, 'percent': percents.round(2)}))

for col in ['gender', 'age', 'time_in_hospital', 'age_numeric']:
  print(f'--- {col} ---')
  show_category_distribution(col)
  print()
```

```python
# Visualization
plt.figure(figsize=(8,6))
sns.barplot(x='gender', y='time_in_hospital', data=df)
plt.title('Average Time in \nHospital by Gender')
plt.ylabel('Average Time in Hospital(days)')
plt.xlabel('Gender')
plt.show()
```

```python
# Visualization 2
df['age_group'] = pd.cut(df['age_numeric'], bins=[5,15,25,35,45],
                          labels=['0-10', '20-30', '30-40', '40+'])

plt.figure(figsize=(8,6))
sns.boxplot(x='age_group', y='time_in_hospital', data=df)
plt.title('Time in Hospital by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Time in Hospital (days)')
plt.show()
```

##In conclusion, gender does not differentiate the amount of time spent in the hospital because, on average, males and females both spent roughly 4 days hospitalized. However, age makes a difference: patients between 80-90 years old spent the most time in the hospital (4.8 days) compared to patients aged 0-10 who spent less time in the hospital (2.5 days) . Therefore, age appears to be a better predictor of hospitalization length for diabetic patients.
