```python
# Import Python Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 100)
pd.set_option('display.float_format', lambda x: f"{x:,.2f}")
pd.set_option('display.width',1000)

plt.style.use('default')
sns.set(style='whitegrid')
```

```python
# Upload csv file
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
#Data Overview
print('\nList first 5 rows')
print(df.head())
print('\nStatistical Description')
print(df.describe())
print(df.info())
print('\nShape (rows, columns):', df.shape)
print('\nColumn names:')
print(df.columns)
# I noticed that there are 's' values 'Mean Score', 'Number Scoring Below 65', 'Percent Scoring Below 65', 'Number Scoring 65 or Above', 'Percent Scoring 65 or Above', 'Number Scoring 80 or Above', 'Percent Scoring 80 or Above' which might cause problem later on.
```

```python
print(df.tail())
# It does look similar to the first 5 rows in terms of having 's' and 'na' values in the same columns as it was in the beginning of the dataset.
```

```python
df.describe(include = 'all')
# I retrieved a statistical summary for each column, it's showing NaNs as results which could be because of the 's' in each column.
```

```python
# Missing Values Analysis
print('\nMissing values in columns:')
print(df.isna().sum())
print('\nPercentage of missing values per column:')
print((df.isna().mean()*100).round(2))
print('\nNumber of unique values per column:')
print(df.nunique())
```

```python
# Random sample of dataset
df.sample(5, random_state=14)
```

```python
# Clean column names
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(' ', '_')
      .str.replace('-', '_')
)
df.columns
```

```python
# Inspect unique values in number_scoring_below 65
df['number_scoring_below_65'].unique()[:50]
```

```python
# Inspect unique values in number_scoring_65_or_above
df['number_scoring_65_or_above'].unique()[:50]
```

```python
# Inspect unique values in number_scoring_80_or_above
df['number_scoring_80_or_above'].unique()[:50]
```

```python
# Replace 's' with actual NaN and convert to numeric
df['number_scoring_below_65'] = df['number_scoring_below_65'].replace('s', np.nan)
df['number_scoring_below_65'] = pd.to_numeric(df['number_scoring_below_65'], errors='coerce')
df['number_scoring_65_or_above'] = df['number_scoring_65_or_above'].replace('s', np.nan)
df['number_scoring_65_or_above'] = pd.to_numeric(df['number_scoring_65_or_above'], errors='coerce')
df['number_scoring_80_or_above'] = df['number_scoring_80_or_above'].replace('s', np.nan)
df['number_scoring_80_or_above'] = pd.to_numeric(df['number_scoring_80_or_above'], errors='coerce')
df[['number_scoring_below_65', 'number_scoring_65_or_above', 'number_scoring_80_or_above']].info()
```

```python
# Check missing values again
df.isna().sum()
```

```python
# Mean imputation for number_scoring_below_65, number_scoring_65_or_above, and number_scoring_80_or_above
below_65_median = df['number_scoring_below_65'].median()
scoring_65_or_above_median = df['number_scoring_65_or_above'].median()
scoring_80_or_above_median = df['number_scoring_80_or_above'].median()
df['number_scoring_below_65'] = df['number_scoring_below_65'].fillna(below_65_median)
df['number_scoring_65_or_above'] = df['number_scoring_65_or_above'].fillna(scoring_65_or_above_median)
df['number_scoring_80_or_above'] = df['number_scoring_80_or_above'].fillna(scoring_80_or_above_median)
df.isna().sum()
```

```python
# Duplicate Rows Analysis
# Number of duplicate rows
num_duplicates = df.duplicated().sum()
print('Number of duplicate rows:', num_duplicates)
# Show duplicate rows (if any)
if num_duplicates > 0:
  print('\nDuplicate rows:')
  display(df[df.duplicated()])
else:
  print('No duplicate rows found.')
# Remove duplicates
df_no_dupes = df.drop_duplicates()
print('Shape after removing duplicates (rows, columns):', df_no_dupes.shape)
```

```python
#Converting the data types of columns: school_dbn, school_name, school_level, and regents exam from object to category.
categorical_cols = ['school_dbn', 'school_name', 'school_level', 'regents_exam']

for col in categorical_cols:
  df[col] = df[col].astype('category')

df[categorical_cols].dtypes
```

```python
# checking to confirm that the data types were changed.
df.info()
```

```python
#Descriptive statistics of numeric columns. I created a new data frame that has only the numeric columns.
numeric_df= df.select_dtypes('float64')
numeric_df.describe()
```

```python
# Individual descriptive statistics summary for each column I'm going to focus on.
print('Scoring Below 65 Summary:')
print(df['number_scoring_below_65'].describe())

print('\nScoring 65 or Above Summary:')
print(df['number_scoring_65_or_above'].describe())

print('\nScoring 80 or Above Summary:')
print(df['number_scoring_80_or_above'].describe())
```

```python
# Selecting a school
school_names = df['school_name'].unique()[:25]
print("School Names:", school_names)
# I select East Side Middle School.
# I'm going to compare East Side Middle School vs all Manhattan Schools.
```

```python
# Displaying the Number Scoring Below 65 for all Manhattan Schools vs East Side Middle School
print('Number scoring below 65 across all Manhattan schools:')
print(df['number_scoring_below_65'].describe())
print('Number scoring below 65 at East Side Middle Scool:')
print(east_side_ms_df['number_scoring_below_65'].describe())
```

```python
# Visualizations:
fig, axes = plt.subplots(1,3, figsize=(18,4))

sns.histplot(df['number_scoring_below_65'], bins=30, kde=True, ax=axes[0])
axes[0].set_title('Number Scoring Below 65 \n Distribution')

sns.histplot(df['number_scoring_65_or_above'], bins=30, kde=True, ax=axes[1])
axes[1].set_title('Number Scoring 65 or Above \n Distribution')

sns.histplot(df['number_scoring_80_or_above'], bins=30, kde=True, ax=axes[2])
axes[2].set_title('Number Scoring 80 or Above \n Distribution')

plt.show()
```
